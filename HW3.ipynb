````python
# Queue class: can be implemented using a list in Python
# front <- x , x , x , <- rear
class Queue:
    def __init__(self, item=0):
        self.fro = None
        self.rear = None
        self.next = None
        self.cnt = 0
        self.item = item

    def enqueue(self, item):
        node = Queue(item)
        self.cnt += 1  

        if (self.fro is None):
            self.fro = node
            self.rear = node
        else:
            self.rear.next = node
            self.rear = node 

    # dequeue()
    def dequeue(self):
        if self.fro is None:
            print('Queue Underflow')
            return
        
        self.cnt-=1
        current = self.fro
        self.fro = self.fro.next 
                
        if self.fro is None:
            self.rear = None    
        
        return current.item
        

    def isEmpty(self):
        if self.fro is self.rear is None:
            return True
        else:
            return False

    def size(self):
        return self.cnt

    def front(self):
        return self.fro.item

    def printall(self):
        
        current = self.fro
        print("front  <-  ",end="" )
        while(current.next != None):
            print(current.item,end=" , ")
            current = current.next
        print(current.item," <- rear")
            

# function for generating all the binary numbers smaller than or equal to a given number n
def generateBinaryNumbers(n):
    
    q = Queue()
    q.enqueue("1")
    tmp = []
    
    while (n > 0):
        s_1 = q.dequeue()
        tmp.append(s_1)
        q.enqueue(s_1+"0")
        s_2 = s_1
        q.enqueue(s_2+"1")
        n-=1
    return tmp

n=10
print("When n is,",n,", the binary numbers are:")
print(generateBinaryNumbers(n))

n=13
print("When n is,",n,", the binary numbers are:")
print(generateBinaryNumbers(n))
````
